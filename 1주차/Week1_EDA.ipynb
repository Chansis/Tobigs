```python
selected_cols = ["Age","MonthlyIncome","OverTime","JobSatisfaction","WorkLifeBalance"]
temp = df[selected_cols]
temp
```

```python
sns.distplot(temp["Age"],hist=True,kde=True,rug=True)
```

```python
Ov_people = temp.loc[temp.OverTime == "Yes",:]
Ov_people
```

```python
Nov_people = temp.loc[temp.OverTime == "No",:]
Nov_people
```

```python
fig, ax = plt.subplots(ncols=2)

a = sns.countplot(x = "JobSatisfaction",data=Ov_people,ax=ax[0])
a.set_title("Overtime")
b= sns.countplot(x = "JobSatisfaction",data=Nov_people,ax=ax[1])
b.set_title("No overtime")
```

```python
fig, ax = plt.subplots(ncols=2)

a = sns.stripplot(x=Ov_people["WorkLifeBalance"],y=Ov_people["MonthlyIncome"],ax=ax[0])
a.set_title("Overtime")
b=sns.stripplot(x=Nov_people["WorkLifeBalance"],y=Nov_people["MonthlyIncome"],ax=ax[1])
b.set_title("No Overtime")
```

```python
heatmap_data = df[["WorkLifeBalance","JobSatisfaction","MonthlyIncome","PercentSalaryHike","YearsAtCompany"]]
colormap = plt.cm.PuBu
plt.figure(figsize=(10, 8))
plt.title("Person Correlation of Features", y = 1.05, size = 15)
sns.heatmap(heatmap_data.astype(float).corr(), linewidths = 0.1, vmax = 1.0,
           square = True, cmap = colormap, linecolor = "white", annot = True, annot_kws = {"size" : 16})
```
